```python
query1 = """
SELECT 
    c.first_name, 
    c.last_name, 
    c.email, 
    a.address, 
    ci.city
FROM 
    customer c
JOIN 
    address a ON c.address_id = a.address_id
JOIN 
    city ci ON a.city_id = ci.city_id
WHERE 
    ci.city_id = 312;
"""
cursor.execute(query1)
result1 = cursor.fetchall()
print("Query 1 Result:")
for row in result1:
    print(row)
```

```python
query2 = """
SELECT 
    f.title, 
    f.description, 
    f.release_year, 
    f.rating, 
    f.special_features
FROM 
    film f
JOIN 
    film_category fc ON f.film_id = fc.film_id
JOIN 
    category c ON fc.category_id = c.category_id
WHERE 
    c.name = 'Comedy';
"""
cursor.execute(query2)
result2 = cursor.fetchall()
print("\nQuery 2 Result:")
for row in result2:
    print(row)
```

```python
query3 = """
SELECT 
    a.last_name, 
    f.title, 
    f.release_year
FROM 
    actor a
JOIN 
    film_actor fa ON a.actor_id = fa.actor_id
JOIN 
    film f ON fa.film_id = f.film_id
WHERE 
    a.first_name = 'Johnny' 
    AND a.last_name = 'Lollobrigida';
"""
cursor.execute(query3)
result3 = cursor.fetchall()
print("\nQuery 3 Result:")
for row in result3:
    print(row)
```

```python
query4 = """
SELECT 
    a.first_name, 
    a.last_name
FROM 
    actor a
JOIN 
    film_actor fa ON a.actor_id = fa.actor_id
JOIN 
    film f ON fa.film_id = f.film_id
WHERE 
    f.title = 'Bingo Talented';
"""
cursor.execute(query4)
result4 = cursor.fetchall()
print("\nQuery 4 Result:")
for row in result4:
    print(row)
```

```python
query5 = """
SELECT 
    p.customer_id, 
    p.amount
FROM 
    payment p
WHERE 
    p.amount > (SELECT 2 * AVG(amount) FROM payment);
"""
cursor.execute(query5)
result5 = cursor.fetchall()
print("\nQuery 5 Result:")
for row in result5:
    print(row)
```

```python
query6 = """
SELECT 
    c.first_name, 
    c.last_name, 
    COUNT(*) AS num_payments
FROM 
    customer c
JOIN 
    payment p ON c.customer_id = p.customer_id
GROUP BY 
    c.customer_id
ORDER BY 
    num_payments DESC
LIMIT 5;
"""
cursor.execute(query6)
result6 = cursor.fetchall()
print("\nQuery 6 Result:")
for row in result6:
    print(row)
```
